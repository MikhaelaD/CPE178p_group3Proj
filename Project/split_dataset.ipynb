```python
import os
import shutil

original_dataset_dir = 'C:/Users/Azra/.vscode/dataset/Chula-ParasiteEgg-11/Chula-ParasiteEgg-11/data' 
split_dataset_dir = 'dataset'

if not os.path.exists(split_dataset_dir):
    os.makedirs(split_dataset_dir)

for filename in os.listdir(original_dataset_dir):
    if '_' in filename:
        class_name = filename.split('_')[0] 
        class_folder = os.path.join(split_dataset_dir, class_name)

        if not os.path.exists(class_folder):
            os.makedirs(class_folder)
        
        src_file_path = os.path.join(original_dataset_dir, filename)
        dst_file_path = os.path.join(class_folder, filename)
        
        shutil.move(src_file_path, dst_file_path)
        print(f"Moved {filename} to {class_folder}")

print("Dataset splitting complete!")
```
